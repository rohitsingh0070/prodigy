```python


import pandas as pd
import matplotlib.pyplot as plt
import os

# Create a folder structure to store the file and outputs
output_folder = "population_data_visualization"
os.makedirs(output_folder, exist_ok=True)

# Load the dataset
# Replace this with the actual path to the dataset file if downloaded
data_url = "https://data.worldbank.org/indicator/SP.POP.TOTL"
# Simulating data loading for demonstration purposes
# In a real scenario, you'd load the dataset using pd.read_csv("path_to_file.csv")

data = {
    "Country": ["India", "China", "USA", "Indonesia", "Pakistan"],
    "Year": [2022, 2022, 2022, 2022, 2022],
    "Population": [1417173173, 1412127053, 334805269, 277534122, 240485658],
}
df = pd.DataFrame(data)

# Plotting the data: Bar chart for population distribution across countries
plt.figure(figsize=(10, 6))
plt.bar(df["Country"], df["Population"], color='skyblue')
plt.title("Population Distribution by Country (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()

# Save the figure
chart_path = os.path.join(output_folder, "population_distribution.png")
plt.savefig(chart_path)

# Display the plot
plt.show()

# Save the dataset for reference
data_file_path = os.path.join(output_folder, "population_data.csv")
df.to_csv(data_file_path, index=False)

print(f"Chart saved at: {chart_path}")
print(f"Dataset saved at: {data_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create a folder structure to store the file and outputs
output_folder = "titanic_data_analysis"
os.makedirs(output_folder, exist_ok=True)

# Load the Titanic dataset
# Replace this with the actual path to the Titanic dataset if downloaded
data_url = "https://www.kaggle.com/c/titanic/data"
# Simulating data loading for demonstration purposes
# In a real scenario, you'd load the dataset using pd.read_csv("path_to_file.csv")

data = {
    "PassengerId": [1, 2, 3, 4, 5],
    "Survived": [0, 1, 1, 1, 0],
    "Pclass": [3, 1, 3, 1, 3],
    "Sex": ["male", "female", "female", "female", "male"],
    "Age": [22, 38, 26, 35, 35],
    "SibSp": [1, 1, 0, 1, 0],
    "Parch": [0, 0, 0, 0, 0],
    "Fare": [7.25, 71.2833, 7.925, 53.1, 8.05],
    "Embarked": ["S", "C", "S", "S", "S"]
}
df = pd.DataFrame(data)

# Data Cleaning
# Checking for missing values
df.isnull().sum()

# Filling missing values (example: filling missing age with median)
df["Age"].fillna(df["Age"].median(), inplace=True)

# Data Analysis
# Overview of the dataset
summary = df.describe()

# Visualizations
# 1. Survival Count
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Survived", palette="Set2")
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "survival_count.png"))
plt.show()

# 2. Survival by Class
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Pclass", hue="Survived", palette="Set1")
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "survival_by_class.png"))
plt.show()

# 3. Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], kde=True, bins=10, color="blue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "age_distribution.png"))
plt.show()

# Save the cleaned dataset and summary
cleaned_data_path = os.path.join(output_folder, "cleaned_titanic_data.csv")
df.to_csv(cleaned_data_path, index=False)

summary_path = os.path.join(output_folder, "data_summary.csv")
summary.to_csv(summary_path)

print(f"Cleaned dataset saved at: {cleaned_data_path}")
print(f"Summary saved at: {summary_path}")
```
